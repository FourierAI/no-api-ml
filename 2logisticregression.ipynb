```python
def sigmoid_fun(x):
    return 1/(1+np.exp(-x))
```

```python
print(sigmoid_fun(2))
```

```python
print(sigmoid_fun(np.array([1,32,5,-1,-20,0])))
```

```python
np.set_printoptions(suppress=True)
```

```python
def feedforward_fun(Theta,X):
#     Theta and X must be matrix (np.array().shape = 2,1)
    rows = np.shape(X)[0]
    Z = np.array([np.dot(Theta,X[i]) for i in range(rows)])
    return sigmoid_fun(Z)
```

```python
feedforward_fun(np.array([1,1]),np.array([[1,2],[5,3]]))
```

```python
def cost_fun(Theta, X, Y_label):
    h_q = feedforward_fun(Theta, X)
    
#     print('h_q shape:',h_q.shape)
#     print('h_q:',h_q)
    cost = Y_label*np.log(h_q)+ (1-Y_label)*np.log(1-h_q)
    return -np.sum(cost)/X.shape[0]
```

```python
print(cost_fun(np.array([1,2,1,1]),np.array([[1,1,1,1],[2,3,2,1]]),np.array([0,1])))
```

```python
sample_indexs = np.random.choice(iris2_data.shape[0],iris2_data.shape[0]//2)
```

```python
train_data = iris2_data[sample_indexs] 
```

```python
print(train_data[:,4][:5])
print(train_data[:,:4][:5])
```

Training process

```python
if __name__ == '__main__':
    
    
    Theta = np.array([0.0,0.0,0.0,0.0])
    
    X = train_data[:,:4]
    Y_label = train_data[:,4]
    learning_rate = 0.01
    dirac_x = 0.0001
    
    diff = 100000
    previous_cost = 0
    while True:
        Theta_list = []
        current_cost = cost_fun(Theta,X,Y_label)
        print('current_cost:',current_cost)
        if abs(previous_cost - current_cost)<=0.00001:
            break
        for i in range(Theta.shape[0]):
#             compute gradiant
            Theta_cp = Theta
            Theta_cp[i] = Theta_cp[i] + dirac_x
            cost_lambda = cost_fun(Theta_cp,X,Y_label)
            gradiant = (cost_lambda-current_cost)/dirac_x
            theta_tmp = Theta[i] - learning_rate*gradiant
            Theta_list.append(theta_tmp)
            
        previous_cost = current_cost
        Theta = np.array(Theta_list)
        print('current Theta:',Theta)
    print('training is over!')
    print('optimal parameters are:',Theta)
```
